```python
!pip install ebooklib requests beautifulsoup4

import requests
from bs4 import BeautifulSoup
from ebooklib import epub

# ---- å¡«å…¥ä½ è¦æŠ“çš„å…¬ä¼—å·æ–‡ç« é“¾æ¥ä»¬ ----
urls = [
    "https://mp.weixin.qq.com/s/k_yo0LLNFvaf2exIBYFUhw",
]

def fetch_content(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
    }
    r = requests.get(url, headers=headers)
    r.encoding = 'utf-8'

    soup = BeautifulSoup(r.text, 'html.parser')
    title = soup.find('h1').get_text(strip=True)
    content = soup.find(id='js_content')

    html = f"<h1>{title}</h1>{str(content)}"
    return title, html

# åˆ›å»º EPUB
book = epub.EpubBook()
book.set_identifier('wx')
book.set_title('å¾®ä¿¡å…¬ä¼—å·åˆé›†')
book.set_language('zh')

chapters = []

for url in urls:
    title, html = fetch_content(url)
    chapter = epub.EpubHtml(title=title, file_name=f"{title}.xhtml", lang='zh')
    chapter.content = html
    book.add_item(chapter)
    chapters.append(chapter)

# è®¾ç½®é¦–é¡µç›®å½•
book.toc = chapters
book.spine = ['nav'] + chapters

# å¿…éœ€èµ„æº
book.add_item(epub.EpubNcx())
book.add_item(epub.EpubNav())

# è¾“å‡º EPUB æ–‡ä»¶
epub.write_epub('wechat_book.epub', book)
print("ğŸ“š å·²ç”Ÿæˆç”µå­ä¹¦ï¼šwechat_book.epub")
```
